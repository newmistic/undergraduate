```python
import re  # 用于正则表达式模式匹配
import requests  # 用于发起HTTP请求

url = "http://guba.eastmoney.com/list,zssh000001.html"
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/92.0.4515.131 Safari/537.36 SLBrowser/8.0.1.4031 SLBChan/105'
}
comments_urls = []
# 发送HTTP GET请求到指定的URL，并使用定义的头部信息
response = requests.get(url, headers=headers)

pattern = re.compile(r'<a.*?href="/(.*?)"')

"""
re.compile() 是 Python 中正则表达式模式编译的函数。它的主要作用是将正则表达式模式编译成一个可重复使用的正则表达式对象。这个对象可以用于多次执行匹配操作，提高了匹配效率，尤其在需要多次使用相同模式的情况下非常有用。

1.	<a：这部分匹配字符串中的<a，表示HTML中的锚标签的开始。
2.	.*?：这是一个非贪婪匹配，表示匹配任意数量（包括零个）的任意字符，直到接下来的模式匹配成功。
3.	href="/：这部分匹配字符串中的href="，表示HTML中的链接属性的开始。
4.	(.*?)：这是一个捕获组，用于匹配任意数量（包括零个）的任意字符，并将其保存在捕获组中。这个捕获组用于匹配URL。
5.	"：这部分匹配字符串中的双引号，表示HTML中的属性值的结束。
因此，这个正则表达式的目标是匹配HTML文本中的<a标签的href属性值，该属性值通常包含了链接的URL。捕获组 (.*?) 用于捕获这些URL，并将它们提取出来，以便后续的处理。这允许代码提取网页中所有链接的URL。
"""

# 使用正则表达式模式在响应的HTML内容中查找所有匹配项
matches = pattern.findall(response.text)

# 初始化一个空列表，用于存储过滤后的匹配项
filtered = []

# 遍历匹配项并过滤掉不需要的URL
for match in matches:
    if "i.eastmoney.com/" not in match:
        filtered.append(match)

# 初始化一个空列表，用于存储最终结果
result = []

# 遍历过滤后的URL，并构建完整的URL
for match in filtered:
    if "/caifuhao.eastmoney.com/" in match:
        url = "http://" + match
    else:
        url = "http://guba.eastmoney.com/" + match

    result.append(url)

# 打印最终的URL列表
print(matches)
print(filtered)
print(result)
```

```python
import re
import requests


def download_page(url, para=None):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/92.0.4515.131 Safari/537.36 SLBrowser/8.0.1.4031 SLBChan/105'}
    if para:
        response = requests.get(url, params=para, headers=headers)
    else:
        response = requests.get(url, headers=headers)
    response.encoding = response.apparent_encoding
    if response.status_code == 200:
        return response.text
    else:
        print("failed to download the page")


url_ = "http://guba.eastmoney.com/list,zssh000001.html"

comments_urls = []
#pattern = r'<a.*?href="(.*?)"'
pattern = re.compile(r'<div.*?class="title".*?a.*?href="(.*?)"')
str1 = download_page(url_)
commons_urls = pattern.findall(str1)

#commons_urls = re.findall(pattern, str1, re.S)
for i in commons_urls:
    print(i)
```
