## part two (50 points)
Download comments with the function download_imdb_comments, and clean the downloaded comments with the function clean. Finnaly, save the cleaned comments in comments.txt. 

* I already offer you the function clean, run it and call it if you want to clean comments
* save clean data in comments.txt, please try [writelines()函数](https://www.runoob.com/python/file-writelines.html)
* please submit the jupyter notebook, you are **not required** to submit the comments.txt

```python
import requests
import re
def download_imdb_comments():
    url = "https://www.imdb.com/title/tt0110912/reviews"
    para = {"ratingFilter":"1"}
    r= requests.get(url,params=para)
    html = r.text
    pattern_comment = r'<div class="text show-more__control">(.*)</div>'
    comments = re.findall(pattern=pattern_comment,string=html)
    return comments
```

```python
# 下载评论
comments = download_imdb_comments()
comments
```

```python
# 去掉html字符,替换&quot &#39字符，&quot;表示空格 ，&#39表示'
def clean(text):
    re_tag = re.compile(r'<[^>]+>') 
    pa_clean_not = re.compile("&#39;")
    pa_clean_quotes = re.compile("&quot;")

    text = re_tag.sub('', text)
    text = pa_clean_not.sub("'",text)
    text = pa_clean_quotes.sub('',text)
    return text
```

**task one - call function clean to clean all the comments**

```python
# 使用clean函数清洗评论
# Please start writing program here
cleaned_comments = []
comments = download_imdb_comments()
for i in comments:
    cleaned_comments.append(clean(i))
print(cleaned_comments)
```

```python
# 在每条评论后添加回车符,您可以根据自己的实际编程需求决定是否使用下一条语句
clean_comments = [c+"\n" for c in cleaned_comments]
clean_comments
```

**task two - save the cleaned comments in comments.txt**

```python
# 清洗后的数据保存到comments.txt中
# 如果保存出现编码问题，尝试在open函数中添加参数encoding='utf-8'
# Please start writing program
with open("comments.txt", "r+", encoding="utf-8") as f:
    f.write("\n".join(clean_comments))
with open("comments.txt", "r+", encoding="utf-8") as f:
    contain = f.read()
print(contain)
```
