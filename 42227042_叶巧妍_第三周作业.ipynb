## 一、字典操作练习 - 字典的创建及基本统计操作

### 1、创建一个学生信息字典列表包括如下内容：

'name':'小张','age':18,'score':77,'ID':'4203357','gender':'男'  
'name':'小李','age':19,'score':86,'ID':'4203358','gender':'男'     
'name':'小红','age':20,'score':72,'ID':'4203314','gender':'女'  
'name':'小赵','age':16,'score':57,'ID':'4203329','gender':'女'  
'name':'小王','age':17,'score':97,'ID':'4203318','gender':'不明'

```python
dic1 = {'name': '小张', 'age': 18, 'score': 77, 'ID': '4203357', 'gender': '男'}
dic2 = {'name': '小李', 'age': 19, 'score': 86, 'ID': '4203358', 'gender': '男'}
dic3 = {'name': '小红', 'age': 20, 'score': 72, 'ID': '4203314', 'gender': '女'}
dic4 = {'name': '小赵', 'age': 16, 'score': 57, 'ID': '4203329', 'gender': '女'}
dic5 = {'name': '小王', 'age': 17, 'score': 97, 'ID': '4203318', 'gender': '不明'}
list_dic_all = [dic1, dic2, dic3, dic4, dic5]
for i in list_dic_all:
    print(f"{i}\n")
```

### 2、学生相关信息基本统计操作

#### （1）请统计女生人数


```python
count = 0
for i in list_dic_all:
    if i['gender'] == '女':
        count += 1
print(f"女生人数为：{count}")
```

#### （2）请输出学号尾数为8的学生姓名


```python
for i in list_dic_all:
    if i['ID'][-1] == '8':
        print(f"{i['name']}")
```

#### （3）请统计成绩高于85分学生人数，打印其姓名和对应成绩

```python
for i in list_dic_all:
    if i['score'] > 85:
        print(f"{i['name']}的成绩为： {i['score']}")
```

#### （4）将列表按照学生成绩升序排序

```python
def take_score(diction):
    return diction['score']

list_dic_all.sort(key=take_score)
for i in list_dic_all:
    print(f"{i}\n")
```

#### 2、统计课程总人数，并打印出不在学生信息表中的学生

```python
print(f"音乐课：{len(music)}人\n数学课：{len(math)}人\n英语课：{len(english)}人\n")

list_class = [music, math, english]
list_dic_name = [dic1['name'], dic2['name'], dic3['name'], dic4['name'], dic5['name'], ]

flag = 0
print("不在学生信息表中的学生：", end="")
for i in list_class:
    for j in range(len(i)):
        if list_dic_name.count(i[j]) == 0:
            print(f"{i[j]}", end=" ")
            flag += 1

if flag == 0:
    print("无")
```

#### 3、分别打印只选了一门课，选了二门课，三门课都选了的学生姓名
