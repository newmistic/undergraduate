```python
# 上证指数吧第一页的url 
pages = []
url = "http://guba.eastmoney.com/list,zssh000001_1.html"
# 请开始编程下载上证指数吧前10页的html内容，每一页内容作为一个元素保存到列表里面

import requests

for i in range(1, 11):
    response = requests.get("http://guba.eastmoney.com/list,zssh000001_{}.html".format(i))
    response.encoding = response.apparent_encoding
    pages.insert(len(pages), response.text)
    print(pages[i-1])
```
