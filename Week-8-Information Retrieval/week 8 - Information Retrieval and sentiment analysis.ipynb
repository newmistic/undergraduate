更多疑问 - “原子能”和”应用”是否应等同对待? 

Based on the analysis of a large number of webpage (text) contents, the more frequently a word appears in all texts, the lower its weight in a specific text

Inverse document frequency - Inverse document frequency (IDF)
$$ IDF(w) = log(\frac{D}{D_w})$$

* $D$ - total number of documents
* $D_w$ - the number of documents containing the term $w$
* 含有这个词的文档越多，这个词的重要性就越低

假定(全世界所有的网页或文本共 有10亿个)（**基于语料库**）

- "原子能"出现在200万个网页中出现
- "的"在10亿个网页中都出现过
- “应用”在5亿个网页中出现

Then tf–idf for a word in a document is calculated as
$$ TF\_IDF(t,d,D) = tf(t,d)* idf(t,D)$$

The relevance of a document based on the query words 
$$TF\_IDF = TF_{w1}*IDF_{w1} + TF_{w2}*IDF_{w2}+…TF_{wn}*IDF_{wn}$$

一篇文章共有1000个词,原子能出现2次, 的 出现35次, 应用 出现5次

```python
text = ['I like programing','I like python, do you like it',  'I like data analysis','data analysis is fun']
```

```python
# extract all the words
all_words = []
for s in text:
    words = re.findall(r'\w+',s)
    all_words.extend(s.split())
```

```python
# create a vocabulary
print(all_words)
voc = set(all_words) # set 的特性，没有重复的元素，变成词袋
voc = list(voc) # 重新变成列表
voc
```

```python
# define a function to create bag of words for a sentence
def create_bag(sentence):
    bag = [] # 空袋子
    words_in_sen = Counter(sentence.split())
    for w in voc:
        if w in words_in_sen:
            bag.append(words_in_sen.get(w)) # 像字典的表达，w的次数添加进去
        else:
            bag.append(0)
    return bag
```

```python
create_bag(text[0])
```

```python
print(voc) #对应 create_bag(text[0])
```

```python
for s in text:
    print(create_bag(s))
```

# sentiment analysis

**Affective states**
* **Emotion**: Relatively brief episode of response to the evaluation of an external or internal event as being of major significance. (angry, sad, joyful, fearful, ashamed, proud, elated, desperate)
* **Mood**: Diffuse affect state, most pronounced as change in subjective feeling, of low intensity but relatively long duration, often without apparent cause. (cheerful, gloomy, irritable, listless, depressed, buoyant)
* **Attitude (Sentiment)**: Relatively enduring, affectively colored beliefs, preferences, and predispositions towards objects or persons. (liking, loving, hating, valuing, desiring)

Golder, S. A., & Macy, M. W. (2011). Diurnal and Seasonal Mood Vary with Work, Sleep, and Daylength Across Diverse Cultures. Science, 333(6051), 1878-1881.

positive and negative affect are independent dimensions. 正面和负面情感相互独立，不能替代
* Positive affect (PA) includes enthusiasm, delight, activeness, and alertness, 
* whereas negative affect (NA) includes distress, fear, anger, guilt, and disgust

```python
sentence = 'We have some delightful new food in the cafeteria. Awesome!'

# 正面情感词表
pos_lst = ['awesome','good','nice','super','fun','delightful']

# 负面情感词表
neg_lst = ['awful', 'lame', 'horrible', 'bad', 'sorrow']
```

```python
cleaned = sentence.lower()
cleaned
```

```python
import string
```

```python
punc = string.punctuation
punc
```

```python
cleaned = "".join([w for w in cleaned if w not in punc])
cleaned
```

```python
wrd_lst = cleaned.split() # 先进行切词
pos_count = len([w for w in wrd_lst if w in pos_lst])
neg_count = len([w for w in wrd_lst if w in neg_lst])
print('The positive score of this text is: ', pos_count/len(wrd_lst))
print('The negative score of this text is: ', neg_count/len(wrd_lst))
```
