```python
def employ_contain(employee_id):
    exist = False
    for e in employee:
        if employee_id == e[0]:
            exist = True
    if exist:
        print ("{0} is in the list".format(employee_id))
    else:
        print("not in the list")
```

tips
- you can download the html documents using the function download_page we learned this week
- you can create a complete url like this [one](http://guba.eastmoney.com/news,zssh000001,900160956.html) by joining "http://guba.eastmoney.com" and "/news,zssh000001,900160956.html".

```python
import re
import requests
def download_page(url):
    try:
        headers = {
        'user-agent': 'Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/52.0.2743.82 Safari/537.36'
        }
        r=requests.get(url,headers=headers)
        if r.status_code==200:
            r.encoding=r.apparent_encoding
            return r.text
    except:
        return "Error"
```
