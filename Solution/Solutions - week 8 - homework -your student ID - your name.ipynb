I offer a function get_sentiment for the sentiment analsyis. When dong the homework, you can call this function if you want.
* function name：get_sentiment
* Input – parameters(texts,poslst,neglst)
* Process – main code block in the function (Data cleaning, calculate PA and NA)
* Output  - return PA and NA
* Test your function

```python
# do not revise any code in this cell
# but you must run it

import string

def get_sentiment(texts,poslst,neglst):
    """this function return the positive and negative score of a text's sentiment"""
    punc = set(string.punctuation)
    new_words = ''.join([w for w in texts.lower() if w not in punc])
    wrd_lst = new_words.split() ### 针对英文文本
    pos_count = len([pos for pos in wrd_lst if pos in poslst])
    neg_count = len([neg for neg in wrd_lst if neg in neglst])
    return pos_count/len(wrd_lst),neg_count/len(wrd_lst)
```

```python
# do not revise any code in this cell
# you can run it
ex = 'We have some delightful new food in the cafeteria. Awesome!!!'
# 正面情感词表
pos_lst = ['awesome','good','nice','super','fun','delightful']
# 负面情感词表
neg_lst = ['awful', 'lame', 'horrible', 'bad', 'sorrow']
# 测试函数
get_sentiment(ex,poslst=pos_lst,neglst=neg_lst)
```

# task 1 - develop a function getPosNegWords （50 points）
>* the function getPosNegWords read and return the sentiment words which saved in neg_file and post_file
>* the parameters are the file paths
>* return the list of postive and negative words

```python
# develop the funciton getPosNegList (50 points)
# tips - remove \n for each positie word and negative word

def getPosNegWords(neg_file,pos_file):
    #please type your code here
    with open(neg_file,"r") as f:
        neg_list = [w.strip() for w in f.readlines()]   ### 去除换行符
        # https://www.runoob.com/python/att-string-strip.html 
    with open(pos_file,"r") as f:
        pos_list = [w.strip() for w in f.readlines()]  
        # Python strip() 方法用于移除字符串头尾指定的字符（默认为空格或换行符）或字符序列。
    return pos_list,neg_list  # 作为返回 
    
    
    
```

# task 2 - develop a function gcalSentimentFile （50 points）
>* the parameter is a file path which contains multiple comemnts, open it, read the file using readlines fuction
>* get the positive/negative list（here call the function getPosNegList）
>* call function get_sentiment to get the sentiment score for each comment in comments
>* return the comments list and the coresponding sentiments list

```python
# develop the function function calSentimentFile (50 points)
def calSentimentFile(file):
    
    
    # please type your code here
    pos_list,neg_list = getPosNegWords(neg_file="negative.txt",pos_file="positive.txt") # 两个变量，接两个内容
    
    with open(file,"r") as f:
        comments = f.readlines()
        
    # get the sentiments
    sentiments =[get_sentiment(texts=c,poslst=pos_list,neglst=neg_list) for c in comments] 
    # 已经写好的函数，每一个元素都是一个帖子
    
    # 列表推导
    return comments,sentiments
    
    
```
