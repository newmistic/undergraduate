# parse html pages with re

```python
import re
import requests
```

```python
def download_page(url,para = None):
    headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/52.0.2743.116 Safari/537.36 Edge/15.15063'}
    if para:
        response = requests.get(url,params=para,headers = headers)
    else:
        response = requests.get(url,headers = headers)
    response.encoding = response.apparent_encoding
    if response.status_code == 200:
        return response.text
    else:
        return None
```

```python
url = "http://guba.eastmoney.com/list,zssh000001.html"
```

```python
html = download_page(url)
```

```python
html
```

```python
pattern = r"""(<div class="articleh.*?">.*?</div>)"""   # 检索方式，一级一级的往下寻找
comments = re.findall(pattern,html,re.S) # 注意换号符
comments
```

```python
len(comments)
```

```python
#comments[0]
```

```python
titles = [] # 任务：取标题
dates = []  # 任务：取反馈的最新时间

title_pattern = r"""<span class="l3 a3">.*?<a.*?href=".*?" title="(.*?)">.*?</a></span>"""
date_pattern = r"""<span class="l5 a5">(.*?)</span>"""

for c in comments:
    
    t = re.findall(title_pattern,c)
    
    if len(t)>0:
        titles.append(t[0])
    else:
        titles.append("")
    
    d = re.findall(date_pattern,c)
    
    if len(d)>0:
        dates.append(d[0])
    else:
        dates.append(0)
```

```python
# The zip() is a built-in Python function. 
# It is used to create an iterator of tuples (known as zip object) from a list of iterables passed as arguments.

for c in zip(titles,dates): ## zip 两个对应的元素组成元组，生成可迭代的变量（对象）
    print (c)
```

enumerate() 函数用于将一个可遍历的数据对象(如列表、元组或字符串)组合为一个索引序列，同时列出数据和数据下标，一般用在 for 循环当中。

enumerate()返回的是一个enumerate对象

```python
# enumerate() allows us to loop over something and have an automatic counter or index.

for index,c in enumerate(zip(titles,dates)):
    print (index,c[0],c[1])
    if index == 10:
        break
```

```python
list(enumerate(seasons, start=1))       # 下标从 1 开始
```

# parse html using BeautifulSoup

```python
from bs4 import BeautifulSoup
import requests
```

```python
# exercise two
def download_page(url,para = None):
    headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/52.0.2743.116 Safari/537.36 Edge/15.15063'}
    if para:
        response = requests.get(url,params=para,headers = headers)
    else:
        response = requests.get(url,headers = headers)
    response.encoding = response.apparent_encoding
    if response.status_code == 200:
        return response.text
    else:
        print ("failed to download the page")  
```

```python
url = "http://guba.eastmoney.com/list,zssh000001.html"
html = download_page(url)
soup = BeautifulSoup(html)
soup
```

* 获取div class = articleh normal_post .*？的评论，其中class属性的字符串可以使用re.compile进行正则化表示
* class_pattern = re.compile("articleh normal_post.*?")

```python
import re
class_pattern = re.compile("listitem.*?")   # compile 编译；编制；编辑；[图情] 汇编  

# 全部取出

# 从compile()函数的定义中，可以看出返回的是一个匹配对象。

all_comments = soup.find_all("tr",{'class':class_pattern}) ##做成变量的形式
#all_comments
```

re.compile 函数

compile 函数用于编译正则表达式，生成一个正则表达式（ Pattern ）对象，供 match() 和 search() 这两个函数使用。

```python
import re
pattern = re.compile(r'\d+')                    # 用于匹配至少一个数字
m = pattern.match('one12twothree34four')        # 查找头部，没有匹配
print(m)
```

```python
m = pattern.match('one12twothree34four', 2, 10) # 从'e'的位置开始匹配，没有匹配
print(m)
```

```python
m = pattern.match('one12twothree34four', 3, 10) # 从'1'的位置开始匹配，正好匹配
print(m)                                       # 返回一个 Match 对象
```

# json

JSON or **JavaScript Object Notation** is a lightweight text-based open standard designed for human-readable data interchange. see the [introduction](https://www.tutorialspoint.com/json/json_overview.htm) 一种数据交换标准

Characteristics of JSON
* JSON is easy to read and write.
* It is a lightweight text-based interchange format.
* JSON is language independent. 独立的计算机语言

JSON supports the following two data structures −
* Collection of **name/value pairs** − This Data Structure is supported by different programming languages.
* Ordered **list** of values − It includes array, list, vector or sequence etc.

JSON（JavaScript Object Notation, JS对象简谱）是一种轻量级的数据交换格式。它基于 ECMAScript（European Computer Manufacturers Association, 欧洲计算机协会制定的js规范）的一个子集，采用完全独立于编程语言的文本格式来存储和表示数据。简洁和清晰的层次结构使得 JSON 成为理想的数据交换语言。 易于人阅读和编写，同时也易于机器解析和生成，并有效地提升网络传输效率。

The following example shows how to use JSON to store information related to books based on their topic and edition.

Json 是字符串格式

"{
   "book": [
	
      {
         "id":"01",
         "language": "Java",
         "edition": "third",
         "author": "Herbert Schildt"
      },
	
      {
         "id":"07",
         "language": "C++",
         "edition": "second",
         "author": "E.Balagurusamy"
      }
   ]
}"

JSON encoder and decoder in python

```python
import json

# encoding to JSON objects

books = { "book": [

  {
     "id":"01",
     "language": "Java",
     "edition": "third",
     "author": "Herbert Schildt"
  },

  {
     "id":"07",
     "language": "C++",
     "edition": "second",
     "author": "E.Balagurusamy"
  }
] }

json.dumps(books)   # 将python对象，编译成 json 字符串形式
# type(books)
```

```python
j = json.dumps(books, sort_keys=True,
                  indent=4, separators=(',', ': '))

print(j)

type(j)
```

```python
# Decode the JSON string to python object like dict

# 将Json字符串解码成python对象

books_dict = json.loads(j)
type(books_dict)
books_dict
```
