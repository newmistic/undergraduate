### 局部变量与全局变量

```python
# Function definition is here ## 局部变量与全局变量是分开的

def changeme(mylist):
    mylist = [1,2,3,4]
    mylist.append([1,2,3,4])
    print ("Values inside the function: ", mylist)## 函数内部以局部变量为优先，打印局部变量（代码级别，生命周期）
    # return mylist

# Now you can call changeme function
mylist = [10,20,30]
changeme( mylist)
print ("Values outside the function: ", mylist) ## 函数外部操作的是全局变量
```

```python
# Function definition is here ## 局部变量与全局变量是分开的

def changeme(mylist):
    mylist.append([1,2,3,4])
    mylist = [1,2,3,4]
    print ("Values inside the function: ", mylist)## 函数内部以局部变量为优先，打印局部变量（代码级别，生命周期）
    # return mylist

# Now you can call changeme function
mylist = [10,20,30]
changeme( mylist)
print ("Values outside the function: ", mylist) ## 函数外部操作的是全局变量
```

一般在函数体外定义的变量成为全局变量，在函数内部定义的变量称为局部变量。
**在函数内部定义的变量，它的作用域也仅限于函数内部，出了函数就不能使用了，我们将这样的变量称为局部变量（Local Variable）。**
全局变量所有作用域都可用，局部变量只能在本函数可用，变量的使用顺序是，局部变量 > 全局变量, 也就是说：优先使用局部变量。
**如果想在函数内使用全局变量，或改变全局变量的值， 应该怎么做呢？**

### 标头参数添加

```python
# headers里的User-Agent查询

def download_page(url,para = None):
    headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/52.0.2743.116 Safari/537.36 Edge/15.15063'}
    if para:
        response = requests.get(url,params=para,headers = headers)
    else:
        response = requests.get(url,headers = headers)
    response.encoding = response.apparent_encoding
    if response.status_code == 200:
        return response.text
    else:
        print ("failed to download the page")  
```

```python
url = "https://www.imdb.com/title/tt0110912/reviews"
para = {"ratingFilter":"1"}
download_page(url,para)
```

```python
import re
news = r"""<a href="http://news.sina.com.cn/c/nd/2018-03-26/doc-ifysqfnf7185680.shtml" target="_blank">
这位航天员从没上过天 但其故事值得全中国人知道</a><span>(03月26日 10:21)</span>"""
pattern =r"""<a.*?target="_blank">(.*?)</a><span>.*?</span>"""
re.findall(pattern, news,re.S)
```

本来，^只匹配字符串的开头，$只匹配字符串结尾，.不匹配换行符

[re.S](https://www.jianshu.com/p/fe7441d9cc7b)做的事情是: 让.也匹配换行符。因此，在字符串中，包含换行符\n，在这种情况下：如果不使用参数re.S，则只在每一行内进行匹配，如果一行没有，就换下一行重新开始。而使用re.S参数以后，正则表达式会将这个字符串作为一个整体，在整体中进行匹配。

re.M做的事情是: 让^匹配每**行**的开头，$匹配每**行**的结尾 ## 每行作为一个单元



```python
text = """医疗器械涨了这么多
上海电气是不是该启动了
有点伤心啊"""

pattern = "^(.*?)$"  # Match anything from the start to end. 非贪婪匹配

# 让^、$只匹配字符串的开头、结尾, .不匹配换行符
ret1 = re.findall(pattern, text)  
print(ret1)
```

```python
# 让.匹配换行符
ret2 = re.findall(pattern, text, re.S)  
print(ret2)
```

```python
# 让^、$匹配每行的开头、结尾, 非贪婪
ret3 = re.findall(pattern, text,re.M)  
print(ret3)
```
