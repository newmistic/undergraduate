# explore the Titanic data

```python
import pandas as pd
```

```python
data = pd.read_csv("titanic.csv")
```

```python
data.head()
```

```python
groupby_sex = data.groupby(["Sex"])
```

```python
import numpy as np
groupby_sex.agg(np.mean)["Survived"].plot(kind = "bar")
```

```python
groupby_Pclass = data.groupby(["Pclass"])
groupby_Pclass.agg(np.mean)["Survived"].plot(kind = "bar")
```
